```python
# Filter for delivered orders 
pre_crisis_delivered_orders = pre_crisis_orders[pre_crisis_orders['is_cancelled'] == 'N']
crisis_delivered_orders = crisis_orders[crisis_orders['is_cancelled'] == 'N'] 
# Merge orders with delivery time sheet
pre_crisis_delivery = pre_crisis_delivered_orders.merge(df_deliveryPerformance[['order_id', 'actual_delivery_time_mins', 'expected_delivery_time_mins']], on='order_id', how='left')
crisis_delivery = crisis_delivered_orders.merge(df_deliveryPerformance[['order_id', 'actual_delivery_time_mins', 'expected_delivery_time_mins']], on='order_id', how='left')

# Calculate average delivery time
avg_delivery_pre = pre_crisis_delivery['actual_delivery_time_mins'].mean()
avg_delivery_crisis = crisis_delivery['actual_delivery_time_mins'].mean()

# Calculate SLA compliance: % of orders delivered within or equal to SLA
sla_compliance_pre = (pre_crisis_delivery['actual_delivery_time_mins'] <= pre_crisis_delivery['expected_delivery_time_mins']).mean() * 100
sla_compliance_crisis = (crisis_delivery['actual_delivery_time_mins'] <= crisis_delivery['expected_delivery_time_mins']).mean() * 100

print(f"Pre-crisis average delivery time: {avg_delivery_pre:.2f} mins")
print(f"Crisis average delivery time: {avg_delivery_crisis:.2f} mins")
print(f"Pre-crisis SLA compliance: {sla_compliance_pre:.2f}%")
print(f"Crisis SLA compliance: {sla_compliance_crisis:.2f}%")
```

```python
# Extract year and month for grouping
df_ratings['year_month'] = df_ratings['review_timestamp'].dt.to_period('M')

# Calculate average rating per month
monthly_avg = df_ratings.groupby('year_month')['rating'].mean()

print(monthly_avg)

# Identify months with sharpest drops by calculating month-over-month difference
monthly_diff = monthly_avg.diff()

# Find months with largest negative changes
sharpest_drops = monthly_diff.nsmallest(3)

print("Months with sharpest drops in average rating:")
print(sharpest_drops)
```

```python
# Filter for crisis period reviews
crisis_reviews = df_ratings[(df_ratings['review_timestamp'] >= '2025-06-01') & (df_ratings['review_timestamp'] <= '2025-09-30')]['review_text'].dropna().astype(str)

# Combine all review texts into one string
text = ' '.join(crisis_reviews).lower()

# Clean up text (remove punctuation)
text = re.sub(r'[^\w\s]', ' ', text)

# Tokenize
words = text.split()

# Remove stopwords (such as "the", "is", "and", "of", "to")
stop_words = set(stopwords.words('english'))
words = [word for word in words if word not in stop_words]

# Define list of typical negative keywords
negative_keywords = ['bad', 'poor', 'late', 'horrible', 'cold', 'terrible', 'worst', 'stale', 'issue','hygiene']

# Count word frequencies
word_counts = Counter(words)
negative_counts = {word: word_counts.get(word, 0) for word in negative_keywords}

# Sort negative keywords by frequency
negative_counts_sorted = dict(sorted(negative_counts.items(), key=lambda item: item[1], reverse=True))

print("Most frequent negative keywords during crisis period:")
print(negative_counts_sorted)
```

```python
# Calculate net revenue (total - discount + delivery fee) for each period
pre_crisis_revenue = (pre_crisis_orders['subtotal_amount'] - pre_crisis_orders['discount_amount'] + pre_crisis_orders['delivery_fee']).sum()
crisis_revenue = (crisis_orders['subtotal_amount'] - crisis_orders['discount_amount'] + crisis_orders['delivery_fee']).sum()

# Calculate revenue loss
revenue_loss = pre_crisis_revenue - crisis_revenue

print(f"Pre-crisis revenue: ₹{pre_crisis_revenue:,.2f}")
print(f"Crisis revenue: ₹{crisis_revenue:,.2f}")
print(f"Estimated revenue loss: ₹{revenue_loss:,.2f}")
```

```python
# Customers with 5 or more orders pre-crisis
pre_crisis_customer_counts = pre_crisis_orders.groupby('customer_id')['order_id'].nunique()
active_customers = pre_crisis_customer_counts[pre_crisis_customer_counts >= 5].index

# Customers ordering during crisis
crisis_customers = crisis_orders['customer_id'].unique()

# Customers who stopped ordering during crisis
stopped_customers = set(active_customers) - set(crisis_customers)

# Get average rating per customer
avg_ratings = df_ratings.groupby('customer_id')['rating'].mean()

# Among stopped customers, count how many have average rating above 4.5
high_rating_stopped = [c for c in stopped_customers if avg_ratings.get(c, 0) > 4.5]

# Print results
print(f"Number of customers with >=5 pre-crisis orders who stopped ordering during crisis: {len(stopped_customers)}")
print(f"Number of these customers with average rating > 4.5: {len(high_rating_stopped)}")
```

```python
# Total spend pre-crisis and crisis by customer
total_spend_pre_crisis = pre_crisis_orders.groupby('customer_id')['subtotal_amount'].sum()
total_spend_crisis = crisis_orders.groupby('customer_id')['subtotal_amount'].sum()

# Identify top 5% high-value customers
threshold = total_spend_pre_crisis.quantile(0.95)
top_customers = total_spend_pre_crisis[total_spend_pre_crisis >= threshold].index

# Order count pre-crisis and crisis for top customers
order_counts_pre = pre_crisis_orders[pre_crisis_orders['customer_id'].isin(top_customers)].groupby('customer_id')['order_id'].nunique()
order_counts_crisis = crisis_orders[crisis_orders['customer_id'].isin(top_customers)].groupby('customer_id')['order_id'].nunique()

# Ratings pre-crisis and crisis for top customers
ratings_pre = df_ratings[(df_ratings['review_timestamp'] >= '2025-01-01') & (df_ratings['review_timestamp'] <= '2025-05-31') & (df_ratings['customer_id'].isin(top_customers))]
ratings_crisis = df_ratings[(df_ratings['review_timestamp'] >= '2025-06-01') & (df_ratings['review_timestamp'] <= '2025-09-30') & (df_ratings['customer_id'].isin(top_customers))]

avg_rating_pre = ratings_pre.groupby('customer_id')['rating'].mean()
avg_rating_crisis = ratings_crisis.groupby('customer_id')['rating'].mean()

# Calculate drops
order_drop = (order_counts_pre - order_counts_crisis).fillna(0)
rating_drop = (avg_rating_pre - avg_rating_crisis).fillna(0)

# Combine into summary
summary = pd.DataFrame({
    'order_drop': order_drop,
    'rating_drop': rating_drop
}).fillna(0)

# Select top customers with largest drops (both order and rating drops)
largest_drop_customers = summary[(summary['order_drop'] > 0) & (summary['rating_drop'] > 0)].sort_values(['order_drop', 'rating_drop'], ascending=False).head(20)

# Explore common patterns: location, cuisine, delivery delays

# Merge orders for these customers with restaurant data
relevant_orders = df_orders[df_orders['customer_id'].isin(largest_drop_customers.index)]
relevant_orders = relevant_orders.merge(df_restaurant[['restaurant_id', 'city', 'cuisine_type']], on='restaurant_id', how='left')

# Check most common city and cuisine for these customers
common_city = relevant_orders.groupby('city')['order_id'].count().sort_values(ascending=False).head(5)
common_cuisine = relevant_orders.groupby('cuisine_type')['order_id'].count().sort_values(ascending=False).head(5)

# Delivery delays: average delivery time for their orders during crisis
delivery_merged = df_deliveryPerformance.merge(relevant_orders[['order_id']], on='order_id', how='inner')
avg_delivery_time = delivery_merged['actual_delivery_time_mins'].mean()
# Create a new column 'delivery_status' with 'On-time' or 'Delayed' values
delivery_merged['delivery_status'] = delivery_merged.apply(
    lambda row: 'On-time' if row['actual_delivery_time_mins'] <= row['expected_delivery_time_mins'] else 'Delayed',
    axis=1
)
delivery_status_counts = delivery_merged['delivery_status'].value_counts()



print("Top 5 cities among the high-value customers with largest drops:")
print(common_city)

print("Top 5 cuisines among the high-value customers with largest drops:")
print(common_cuisine)

print(f"Average delivery time for the high-value customers' orders during crisis: {avg_delivery_time:.2f} mins")
print("Delivery status among the high-value customers with largest drops:")
print(delivery_status_counts)

print("Customers with largest drops:")
print(largest_drop_customers)
```
